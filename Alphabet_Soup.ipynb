## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
appDf = application_df.copy()
appDf.drop(['EIN','NAME'],axis=1,inplace=True)
appDf.head()
```

```python
# Determine the number of unique values in each column
for col in appDf:
  column = appDf[col]
  print(column.name)
  print(len(column.unique()))
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
appDf['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = ['T9','T13','T12','T2','T25','T14','T29','T15','T17']

# Replace in dataframe
for app in application_types_to_replace:
    appDf['APPLICATION_TYPE'] = appDf['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
appDf['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
class_value_counts = appDf['CLASSIFICATION'].value_counts()
class_value_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_value_counts[class_value_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classes = appDf['CLASSIFICATION'].unique()
print(classes)
vals_to_keep =['C1000','C2000','C1200','C3000','C2100']
classifications_to_replace = [i for i in classes if i not in vals_to_keep]

# Replace in dataframe
for cls in classifications_to_replace:
    appDf['CLASSIFICATION'] = appDf['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
appDf['CLASSIFICATION'].value_counts()
```

```python
print(appDf.dtypes)
print(appDf.columns)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
cat_dummies = pd.get_dummies(appDf[['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE',
       'ORGANIZATION','INCOME_AMT', 'SPECIAL_CONSIDERATIONS']])
mergedDf = pd.concat([appDf,cat_dummies],axis=1)
mergedDf.drop(['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE',
       'ORGANIZATION','INCOME_AMT', 'SPECIAL_CONSIDERATIONS'],axis=1,inplace=True)
mergedDf.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = mergedDf.copy()
X.drop('IS_SUCCESSFUL',axis=1,inplace=True)
y = mergedDf['IS_SUCCESSFUL'].values.reshape(-1,1)

# Split the preprocessed data into a training and testing dataset
X_train,X_test,y_train,y_test = train_test_split(X,
                                                 y,
                                                 random_state=97)
print(X_train.shape)
print(X_test.shape)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
num_features = len(X_train_scaled[0])
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=5,activation='relu', input_dim=num_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=5,activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy',
           optimizer='adam',
           metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('AlphabetSoupCharity.h5')
```
